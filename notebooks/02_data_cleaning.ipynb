```python
"""
# 02 - Data Cleaning
Goal: Clean raw text and normalize labels.
"""
```

```python
import pandas as pd
import re
```

```python
data = pd.read_csv(
    r"C:\Users\Moses Tumbo\OneDrive\Documents\model\tic-tac-toe\data\spam.csv",
    encoding='latin-1'
)
data = data.rename(columns={"v1":"label","v2":"text"})
data = data[['label','text']]
```

```python
# --- Define cleaning function ---
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)    # remove links
    text = re.sub(r"\W+", " ", text)       # remove punctuation
    text = re.sub(r"\d+", "", text)        # remove numbers
    text = text.strip()
    return text
```

```python
# --- Apply cleaning ---
data['cleaned_text'] = data['text'].apply(clean_text)
```

```python
# --- Normalize labels ---
data['label'] = data['label'].map({'spam':1, 'ham':0})
```

```python
# --- Drop empty rows ---
data = data[data['cleaned_text'].str.strip().astype(bool)]
```

```python
# Save cleaned dataset back to spam.csv with UTF-8 encoding
import pandas as pd
import os

# Define the save path
save_path = r"C:\Users\Moses Tumbo\OneDrive\Documents\model\tic-tac-toe\data\spam.csv"

# Save the data
data.to_csv(
    save_path,
    index=False,
    encoding='utf-8'
)

print("✅ Saved cleaned dataset with UTF-8 encoding.")
print(f"Saved to: {save_path}")
```
